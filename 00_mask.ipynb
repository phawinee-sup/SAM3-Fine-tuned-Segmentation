```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('IMG_7160.JPG')
image.shape
```

```python
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
# img ต้องเป็น BGR ถ้าอ่านด้วย OpenCV
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# ช่วงสีเขียว (ปรับได้)
lower_green = np.array([35, 40, 80])
upper_green = np.array([85, 255, 255])

mask = cv2.inRange(hsv, lower_green, upper_green)

# แปลงให้เป็น 0 กับ 1
binary_mask = (mask > 0).astype(np.uint8)

plt.imshow(binary_mask, cmap='gray')
plt.title("Binary Mask")
plt.show()
```

```python
import cv2
import numpy as np
import os

def create_green_mask_and_save(input_path, output_folder):
    # อ่านภาพ (BGR)
    image = cv2.imread(input_path)
    
    if image is None:
        raise ValueError(f"ไม่สามารถโหลดภาพได้ กรุณาเช็ค path: {input_path}")

    # แปลงเป็น HSV
    hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # ช่วงสีเขียว (ปรับได้)
    lower_green = np.array([45, 60, 80])
    upper_green = np.array([75, 255, 255])

    # สร้าง mask
    mask = cv2.inRange(hsv, lower_green, upper_green)

    # ทำให้เป็น 0/1
    binary_mask = (mask > 0).astype(np.uint8)

    # ---- เตรียม save ----
    os.makedirs(output_folder, exist_ok=True)

    filename = os.path.basename(input_path)
    name, ext = os.path.splitext(filename)

    output_path = os.path.join(output_folder, f"{name}_mask.png")

    # ตอน save ควรเป็น 0/255 เพื่อให้ดูได้ปกติ
    cv2.imwrite(output_path, binary_mask * 255)

    print(f"บันทึกไฟล์เรียบร้อย: {output_path}")
```

```python
file_list = os.listdir()

for file_name in file_list:
    if file_name.endswith(".JPG") or file_name.endswith(".jpg") or file_name.endswith(".png"):
        create_green_mask_and_save(
            input_path=file_name,
            output_folder="Gaeul_ready"
        )
```
